# Support  Vector Machines

```python
import numpy as np
import pandas as pd
import scipy.io as sio
import scipy.optimize as opt
import seaborn as sns
from matplotlib import pyplot as  plt
from sklearn import svm
```

```python
data = sio.loadmat('/home/leedo/ML_Andrew_Ng/ML-exe/exe6-SVM/data/ex6data1.mat')
data
```

```python
## 可视化数据
df = pd.DataFrame(data['X'],columns=['X1','X2'])
df['y'] = data['y']


#sns.scatterplot('X1','X2',hue = 'y',data = df)
positive = df[df['y'].isin ([1])]
negative = df[df['y'].isin ([0])]

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(positive['X1'],positive['X2'],marker = 'x',label = 'positive')
ax.scatter(negative['X1'],negative['X2'],marker = 'o',label = 'negative')
ax.legend()
plt.show()
```

```python
svc1 = svm.SVC(kernel='linear',C=1, max_iter=1000,tol=0.0001)
#svc1 = svm.LinearSVC(C=1,loss='hinge', max_iter=1000)
svc1
```

```python

svc1.fit(df[['X1','X2']],df['y'])
svc1.score(df[['X1', 'X2']], df['y'])
```

```python
print("svc1.coef_:{}".format(svc1.coef_))
print("svc1.support_vectors_:{}".format(svc1.support_vectors_))
print("svc1.intercept_:{}".format(svc1.intercept_))
```

```python
###绘制分类效果

def drawing(svc):
    sns.scatterplot('X1','X2',hue = 'y',data = df) 
    ##y=a.x +b: w_0x + w_1y +w_3=0 转换为 y = -(w_0/w_1) x - (w_3/w_1)
    w = svc.coef_[0]
    k = -w[0]/w[1]
    x = np.linspace(0,5)
    y = k*x - (svc.intercept_[0])/w[1]
    ##绘制下边界
#    b = svc.support_vectors_[0]
#    y_down = k*x + (b[1] - k*b[0])
    ##绘制上边界
#    b = svc.support_vectors_[-1]
#    y_up = k*x + (b[1] - k*b[0])

    plt.plot(x,y)
#    plt.plot(x,y_up)
#    plt.plot(x,y_down)
    plt.show()
```

```python
drawing(svc1)
```

```python
df['SVM 1 Confidence'] = svc1.decision_function(df[['X1', 'X2']])

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(df['X1'], df['X2'], s=50, c=df['SVM 1 Confidence'], cmap='seismic')
ax.set_title('SVM (C=1) Decision Confidence')
plt.show()
```

# C = 100

```python
svc100 = svm.SVC(kernel='linear',C=100, max_iter=1000,tol=0.0001)
#svc1 = svm.LinearSVC(C=1,loss='hinge', max_iter=1000)
svc100.fit(df[['X1','X2']],df['y'])
svc100.score(df[['X1','X2']],df['y'])
```

```python
drawing(svc100)
```

可以看出相比C=1时，分类更加准确，但是这样回导致泛化性较差

```python
df['SVM 100 Confidence'] = svc100.decision_function(df[['X1', 'X2']])

fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(df['X1'], df['X2'], s=50, c=df['SVM 100 Confidence'], cmap='seismic')
ax.set_title('SVM (C=100) Decision Confidence')
plt.show()
```
