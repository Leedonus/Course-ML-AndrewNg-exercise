## Gaussian distribution

```python
def Gaussian_Distribution(mu,sigma2):
    '''
    Parameters
    ----------
    mu 样本均值
    sigma2: 样本方差
    
    Returns
    -------
    Gaussian  高斯分布概率密度函数
    '''
    
    Gaussian = multivariate_normal(mean=mu, cov=sigma2)
#    P = 1/(np.sqrt(2*np.pi*sigma2))*np.exp(-pow(x-mu,2)/(2*sigma2))
    return Gaussian
```

```python
## 计算均值和方差
mu = np.mean(X,axis = 0)
print(mu)

sigma2 = np.cov(X.T)##以每行为变量
print(sigma2)
M = X.shape[0]
```

```python
Gaussian = Gaussian_Distribution(mu, sigma2)
# 生成二维网格平面
x, y = np.meshgrid(np.linspace(10,20,M), np.linspace(10,20,M))

# 二维坐标数据
d = np.dstack([x,y])
# 计算二维联合高斯概率
z = Gaussian.pdf(d).reshape(M,M)

'''二元高斯概率分布图'''
fig = plt.figure(figsize=(6,4))
ax = Axes3D(fig)
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='seismic', alpha=0.8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
'''二元高斯概率分布图水平面投影'''
plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
x1, y1 = X[:,0],X[:,1]
plt.plot(x1, y1, 'ro', alpha=0.5)
plt.contour(x, y, z, levels=5, alpha =1, zorder=10);
plt.show()
```

```python
##选择阈值
def select_epsilon(X,label):
    '''
    Parameters:
    X： 数据
    label: 标签
    
    return:
    score: F1 score
    epsilon: 阈值
    p_val:高斯概率
    '''
    p_val = Gaussian.pdf(X)
    epsilon_candidate = np.linspace(min(p_val),max(p_val),1200)
    
    score = []
    for e in epsilon_candidate:
        y_pre = (p_val<=e).astype('int')
        score.append(f1_score(label,y_pre))
    return epsilon_candidate[np.argmax(score)],max(score),p_val
```

```python
epsilon , max_score,pval= select_epsilon(X_val,y_val)
epsilon ,max_score
```

```python
y_pre = (pval<=epsilon).astype('int')
ano = X[np.nonzero(y_pre)]
print(ano)
'''二元高斯概率分布图水平面投影'''
plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
x1, y1 = X[:,0],X[:,1]
plt.plot(x1, y1, 'ro', alpha=0.5)
plt.contour(x, y, z, levels=5, alpha =1, zorder=10);
plt.scatter(ano[:,0],ano[:,1])
plt.show()
```
