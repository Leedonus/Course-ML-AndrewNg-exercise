```python
def sigmod(z):
    return 1/(1+np.exp(-z))
```

```python
def cost_data(X,y,theta):
    m  = X.shape[0]
    one = np.ones((m,1))
    predict = sigmod(X.dot(theta))
    first = np.multiply(y,np.log(predict))
    second = np.multiply((one-y),np.log(one-predict))
    return -(1/m)*np.sum(first+second)
```
