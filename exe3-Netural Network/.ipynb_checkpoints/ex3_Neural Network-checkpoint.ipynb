```python
#随机展示一张图
def plot_image():
    rand = np.random.randint(0,5000)
    img = data['X'][rand,:]
    fig,ax = plt.subplots(figsize=(1,1))
    ax.matshow(img.reshape((20,20)).T,cmap=plt.cm.binary)
    plt.show()
#多张可视化
def plot_many_images(X,n):
    data_idx = np.random.choice(np.arange(X.shape[0]),n*n)
    data_example = X[data_idx,:]
    print(data_example.shape)
    fig,ax = plt.subplots(nrows = n,ncols=n,figsize=(8,8))
    #给的数据集数字的朝向都是向右的，因此对矩阵转置
    for i in range(n):
        for j in range(n):
            ax[i,j].matshow(data_example[n*i+j].reshape(20,20).T,cmap=plt.cm.binary)
    plt.show()
```

## 数据处理
对输入添加偏差项，由于数据显示的图像是右向的，进行更改；对标签进行one-hot编码，

```python
#更改输入
x = data['X']
x = np.array([im.reshape((20, 20)).T for im in x])
x = np.array([im.reshape(400) for im in x])
##第一列插入1
x = np.insert(data['X'],0,np.ones(x.shape[0]),axis = 1)

##因为需要一对多的预测，因此需要进行one-hot编码
def one_hot(data):
    y_tmp = []
    for i in range(len(data)):
        tmp = np.zeros(10)
        tmp[data[i]-1] = 1
        y_tmp.append(tmp)
    y_tmp = np.array(y_tmp)
    y = np.concatenate([np.array([y_tmp[:,-1]]).T,np.array(y_tmp[:,:-1])],axis = 1)
    return y.T
```

```python
y = one_hot(data['y'])
print(y.shape)
y
```

## 1维和K维模型预测

逻辑回归代价函数：

$J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{{h}_{\theta }}\left( {{x}^{(i)}} \right) \right)]}$



正则化逻辑回归梯度下降：
\begin{align}
  & Repeat\text{ }until\text{ }convergence\text{ }\!\!\{\!\!\text{ } \\ 
 & \text{     }{{\theta }_{0}}:={{\theta }_{0}}-a\frac{1}{m}\sum\limits_{i=1}^{m}{[{{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}}]x_{_{0}}^{(i)}} \\ 
 & \text{     }{{\theta }_{j}}:={{\theta }_{j}}-a\frac{1}{m}\sum\limits_{i=1}^{m}{[{{h}_{\theta }}\left( {{x}^{(i)}} \right)-{{y}^{(i)}}]x_{j}^{(i)}}+\frac{\lambda }{m}{{\theta }_{j}} \\ 
 & \text{          }\!\!\}\!\!\text{ } \\ 
 & Repeat \\ 
\end{align}

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))


#代价计算函数
def compute_cost(theta,x,y):
    return np.mean(-y*np.log(sigmoid(x.dot(theta)))-(1 - y)*np.log(1-sigmoid(x.dot(theta))))

#梯度计算函数，这里只是计算一个batch 
def gradient(theta,x,y):
    return (1/len(x))*x.T.dot(sigmoid(x.dot(theta))-y)


####正则化
def compute_cost_reg(theta,x,y,lamb):
    return compute_cost(theta,x,y)+ + (lamb/(2*len(x)))*np.sum(pow(theta[1:],2))


def gradient_reg(theta,x,y,lamb):
    grad = np.zeros(x.shape[1])
    grad_tmp = (1/len(x))*x.T.dot(sigmoid(x.dot(theta))-y)
    grad[0]  = grad_tmp[0]  
    grad[1:] =  grad_tmp[1:]+(lamb/len(x))*theta[1:]
    return  grad
```

```python
#单个逻辑回归神经元
def logistic_regression(x,y,lamb):
    theta = np.zeros(x.shape[1])
    
    res = opt.minimize(fun=compute_cost_reg,
                       x0=theta,args=(x,y,lamb),
                       method='TNC',
                       jac=gradient_reg,
                       options={'disp': True})
    
    theta_res = res.x
    return theta_res
```

```python
#预测函数
def predict(x, theta):
    pre = sigmoid(x.dot(theta))
    return (pre >= 0.5).astype(int)
```

```python
t0 = logistic_regression(x, y[0],lamb=1)
```

```python
y_pred = predict(x, t0)
print('Accuracy={}'.format(np.mean(y[0] == y_pred)))
```

## one vs all prediction

```python
#一对多分类
def one_vs_all(x,y,lamb,K=10):
    N = x.shape[1]
    theta_final=np.zeros((N,K))
    for i in range(K):
        theta_final[:,i] = logistic_regression(x,y[i],lamb)
    return theta_final
```

```python
theta_final = one_vs_all(x,y,lamb=1,K=10)
#print(theta_final)
```

```python
predict_prob = x.dot(theta_final)
np.set_printoptions(precision= 25,suppress=True)
predict_prob
```

```python
prod_mat = np.argmax(predict_prob,axis=1)
```

```python
prod_mat
```

```python
data['y'][data['y']==10]=0
y_truth = data['y']
```

```python
print(classification_report(y_truth, prod_mat))
```
